# Cleaning the data so it is easier to use

```python
import pandas as pd
```

## [COVID](https://data.cityofchicago.org/Health-Human-Services/COVID-19-Cases-Tests-and-Deaths-by-ZIP-Code/yhhz-zm2v/data)


```python
df = pd.read_csv('COVID-19_Cases__Tests__and_Deaths_by_ZIP_Code.csv')
df.dropna(subset=['ZIP Code Location'], inplace = True)
longitude = df['ZIP Code Location'].str.replace('POINT \(','').str.split().str[0]
latitude = df['ZIP Code Location'].str.replace('POINT \(','').str.split().str[1].str.replace('\)','')
df['longitude'] = longitude
df['latitude'] = latitude
df.to_csv('COVID-19_Cases__Tests__and_Deaths_by_ZIP_Code_fix.csv', index=False)
df
```
